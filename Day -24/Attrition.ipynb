```python
labelencoder = LabelEncoder()
```

```python
df.head()
```

```python
features = ["Age","BusinessTravel","Department","DistanceFromHome","Education","EducationField","Gender","JobLevel","JobRole",
           "MaritalStatus","MonthlyIncome","NumCompaniesWorked","PercentSalaryHike","StockOptionLevel","TotalWorkingYears","TrainingTimesLastYear","YearsAtCompany","YearsSinceLastPromotion","YearsWithCurrManager"]
```

```python
rf_model = RandomForestClassifier(n_estimators=1000, max_features=2,oob_score=True)
```

```python
features
```

```python
df = df.dropna()
```

```python
rf_model.fit(X = df[features] ,y = df["Attrition"])
```

```python
print("OOb_score",rf_model.oob_score_)
```

```python

for feature,imp in zip(features,rf_model.feature_importances_):

    print(feature,imp)
```

```python
### Decision Tree
```

```python
from sklearn import tree
```

```python
tree_model = tree.DecisionTreeClassifier(max_depth=8)
```

```python
df.head()
```

```python
predictors = pd.DataFrame([df["Age"],df["DistanceFromHome"],df["MonthlyIncome"],df["NumCompaniesWorked"],df["PercentSalaryHike"],df["TotalWorkingYears"],df["YearsAtCompany"],df["JobRole"]]).T
```

```python
predictors.head()
```

```python
tree_model = tree_model.fit(X = predictors,y=df["Attrition"])
```

```python
with open("Atrrition_dt_max_8.dot",'w') as f:
    f = tree.export_graphviz(tree_model,feature_names=["Age","DistanceFromHome","MonthlyIncome","NumCompaniesWorked","PercentSalaryHike","TotalWorkingYears","YearsAtCompany","JobRole"],out_file=f)
```

```python
from sklearn.tree import export_text
r = export_text(tree_model ,max_depth=8 ,feature_names = ["Age","DistanceFromHome","MonthlyIncome","NumCompaniesWorked","PercentSalaryHike","TotalWorkingYears","YearsAtCompany","JobRole"])
```
