```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("train.csv")
```

```python
le = preprocessing.LabelEncoder()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
le.fit_transform(df["Sex"])
```

```python
le.classes_
```

```python
df["Sex"] = le.transform(df["Sex"])
```

```python
df["Embarked"] = le.fit_transform(df["Embarked"])
```

```python
df.head(2)
```

```python
y = df["Pclass"]
```

```python
x = df.drop(["Pclass","Name","PassengerId","Cabin","Ticket"],axis =1)
```

```python
x.head()
```

```python
x_train, x_test, y_train ,y_test = train_test_split(x ,y  , test_size = 0.3, random_state = 3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn import neighbors
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors=3)
```

```python
knn = knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
y_pred =knn.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

### Function to Check knn for function of n neighborabss

```python
def KNN(k):
    knn=neighbors.KNeighborsClassifier(n_neighbors=k)
    knn = knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    print("\n")
    print("accuracy score ",k,":",knn.score(x_test,y_test))
    print("Confusion Matrix ",k,":\n",confusion_matrix(y_test,y_pred))
```

```python
for i in range(1,268):
    KNN(i)
```
